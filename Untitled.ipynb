# Robust Key Information Extraction for Electronic Health Records

## Abstract
Understanding visually rich medical records automatically are a crucial component of Electronic Health Records. Due to the scarcity of medical data and the fact that most approaches in the field of document comprehension are focused on general issues, there are no methods that produce spectacular results for the challenge of extracting key-value information from medical records. In this paper, we provide a brand-new document dataset named VinEHR. There are both orderly tables and unstructured text in the medical records. Additionally, we propose the Multitask Multi-modal Network BROS++, which integrates text, layout, and particularly vision characteristics by the Region of Interest Align operator as significant inputs of the model to concurrently perform the Entity Extraction and Relation Extraction task. Our model achieves significantly better results than the state-of-the-art methods in the VinEHR dataset and benchmark datasets such as FUNSD and SROIE.

# 1. Introduction
Extracting key-value pairs from documents plays an important role in the key information extraction pipeline that automatically explores knowledge from medical records such as personal information and examination results. Such task usually comes after text detection and text recognition in the pipeline, as shown in Figure 1, and aims to assign specific classes to text entities (Entity Extraction-EE) and links between two relevant entities (Relation Extraction-RE). The key-value extraction can save time processing patients’ medical records and extracting indicator results. These results can then be utilized to detect potential health disorders or diseases.

![Figure 1: Illustrations of document KIE pipeline](images/figure_1_kie_pipeline.png)
<p style="text-align: center;">Figure 1: Illustrations of document KIE pipeline </p>


Following the great development of technology, the demand for medical record digitalization or intelligent Electronic Health Records (EHR) has significantly increased. Medical examination records of users can be automatically digitized so that personal information and indicator results are extracted and stored in a short period. This will definitely help the lookup process for users and the development of downstream machine learning (ML) tasks to be more precise. To the best of our knowledge, there is a shortage of public medical examination images dataset despite their importance to the medical domain. Therefore, in this paper, we introduce a new large electrical health records dataset called VinEHR,  which comprises many document form layouts from some of the top medical facilities in Vietnam.

Beside our dataset, we develop a multi-modal network for document image understanding based on BROS [1] network. Our model fully exploits the textual, visual, and positional information of every semantically meaningful component in a document, e.g., text block, heading, and table structure. This model has a spatial Encoder that efficiently extracts relation information of text boxes in 2D space. Vision features are not considered in this model when encoding. Due to this, the method is ineffective when applied to semi-structured datasets, as are other approaches with similar goals, such as LiLT [2] and WYVERN [3]. So, we build a feature extraction module in the encoder to leverage text information such as size, font, and color. In the decoder, we provide a multitask network to learn and infer two tasks, RE and EE, simultaneously. In summary, our main contributions to this paper are:

- We establish a new large Vietnamese Electrical Health Records (VinEHR) dataset for key information extraction. 

- We present an efficient key-value extraction approach from documents that exploit line-level annotations in the dataset and visual knowledge from the image. Moreover, online text augmentation is applied to generalize unseen form structures better.

## 2. Related works

- **Document Image Understanding (DIU)**: Recently published articles have successfully combined image, layout, and text elements to extract information from structurally rich texts. SelfDoc [4], DocFormer [5], LayoutLM [6], StrucText [7], and other methods employ vision features taken from images and concatenate them with textual and layout data, whereas BROS [1] and LiLT [2] concentrate on extracting features from textual and layout. The pre-training of BERT [8] or transformer-encoder [9] models using document pictures has begun in almost all types of projects. Even though these methods have been proven to work in VrDU, They per- form poorly on semi-structured datasets and datasets with a high information density. 

- **Document Understanding Dataset**: A small number of benchmark datasets have been published in several years, notably FUNSD, SROIE, and SciTSR. FUNSD [10] consists of 199 real, fully annotated, scanned forms in English. Numerous financial, accounting and tax forms are included in SROIE [11]. Because some of the text in the dataset is handwritten, this presents a significant challenge. With the help of scientific articles, SciTSR [12] created a massive dataset for table structure identification that includes 15,000 PDF files with the accompanying structure labels. However, there are not many datasets derived from medical records. It makes extracting crucial information from EHR becomes more complex.

## 3. PROPOSED DATASET: VinEHR

### 3.1. Task definition

Tackling the key-value extraction means solving two important sub-tasks: Entity Extraction (EE) and Relation Extraction (RE). The EE task aims to assign a specific type of entity to text blocks, while the RE task finds pairs of entities that are related hierarchically or semantically. The illustration of both tasks is shown in Figure 2, where each text line is given an entity type, and one entity is connected to other entities in the same table row.

![Figure 2: Illustration of EE and RE tasks](images/ee_re_illstr.jpg)
<p style="text-align: center;">Figure 2: Illustration of EE and RE tasks</p>

### 3.2. Dataset construction

**Data collection**: Our new VinEHR dataset is sampled from real-world medical examination records from many renowned hospitals in Vietnam, and each document contains both tabular and textual data. The camera captures them in smartphones in various backgrounds and lighting conditions. Additionally, to test the models’ generalization and robustness, we also crawl documents publicly available on the internet to construct a private test set that includes a strictly table-formed layout but is different from the main dataset (like borderless tables).

![Figure 3: Linking groups and how to label them](images/linking_groups.png)
<p style="text-align: center;">Figure 3: Linking groups and how to label them</p>

**Data processing**: Unlike scanned documents, those captured by the camera often are rotated a certain number of degrees and contain background information. This is irrelevant to document visual information and a burden to the performance of the key information extraction pipeline because the results can be vastly different if documents are put in another unseen background. Therefore, we use a preprocessing pipeline to construct image portions in our VinEHR dataset, which consists of alignment from OCR boxes and perspective transformation from detecting four corners of documents. Next, an OCR tool is applied to these preprocessed images to obtain line-level bounding boxes (like in Fig.6), which human annotators can make use of to label key-value pairs using a labeling tool named AI Scaler 1. They are asked to perform entity labeling, which is deciding box entity type among four classes (header, key, value, and other), then entity linking, that is connecting all boxes in one linking group with a polyline, like in Figure 3. After a few postprocessing steps, the box linking shown in Figure 2 is finally obtained. In addition, we also developed an offline image augmentation tool that randomly replaces texts in scanned documents with relevant contents and applies augmentation techniques while keeping the annotated relations intact

### 4.1 Feature Extraction and Combination

This section presents how to extract document features from the text, images, and bounding boxes. The text encoder and position encoder of BROS++ use BROS architecture. Layout (text placement in 2D space) is recorded as a positional embedding vector, while text input is routed via a pre-trained BERT [8] model into a token embedding vector. Most DIU approaches tokenize all OCR-recognized texts and concatenate them word by word. When utilizing this format for EE or RE problems, every word in the layout area is isolated and has independent outcomes, making model learning difficult. We propose token line level (or paragraph) to input embedding vectors sentence-by-sentence. Our new approach understands whole phrases rather than individual words. All texts are encoded into a text embedding vector Etext with the form (B, hid), where B is the batch size and hid is the hidden size of the BERT [8] model.

Text Layout with 4-vertex polygon format $\mathbf{S_b} \in \mathbb{R}^4$ with shape ($\mathbf{B}$, $\mathbf{N}$, 4) where $\mathbf{N}$ is the number of text boxes detected in image is converted to matrix $\mathbf{S_{re}}$  ($\mathbf{B}$, $\mathbf{N}$, $\mathbf{N}$, 4) where $\mathbf{S_{re_{ij}}}$ is subtraction between $\mathbf{S_i}$ and $\mathbf{S_j}$. This transform to aim for extracting relationships between text layouts. The layout embedding vector is calculated as follows: 
\begin{equation}
    \mathbf{E}_{layout} = \textrm{PE}(\mathbf{X_p}) * \textrm{PE}(\mathbf{Y_p}) \nonumber
\end{equation}

Where $\mathbf{X_p}$ and $\mathbf{Y_p}$ are $\mathbf{1D}$ coordinates extracted from $\mathbf{S}_{re}$ (in the $\mathbf{X}$ and $\mathbf{Y}$ dimensions respectively in $\mathbf{2D}$ space). $\textrm{PE}$ is $\mathbf{1D}$ Position Embedding Encoder layer. Results that are encoded from $\mathbf{X_p}$ and $\mathbf{Y_p}$ by $\textrm{PE}$ are combined to get $\mathbf{2D}$ position Embedding vector.

The vision encoder takes images as inputs and generates a vision embedding vector, including image features for learning models. An image $I$ is initially supplied to the Features Network (FPN) [13], which uses Resnet50 [14] as its main backbone in order to extract features. Normally, the FPN network is used to extract information at multiple scales, but in this case, we use the biggest scale features map as the text is small in the image. The feature map and scaled bounding boxes $B$ will be accepted as inputs by the RoI pooling layer that applies RoI Align [15] as an operator technique. It filters features for each related 2D layout to produce a fixed size embedding vector.

\begin{equation}
    \mathbf{E}_{vision} = \mathcal{L} \left(
       \textrm{RoI} \left(
            \textrm{FPN} \left( \textrm{I} \right), {B_i}
        \right)
    \right) \nonumber
\end{equation}

Where $\mathcal{L}$ is a fully connected layer, transform the vector to 1024 dimensions, and $dim$ is the number of feature maps in FPN output. Before entering the transformer model, all embedding vectors (text embedding vector, text position embedding vector, layout embedding vector, vision embedding vector) representing text, layout, and vision are integrated.
\begin{equation}
    \mathbf{E}_{combine} = \mathbf{E}_{text} +
    \mathbf{E}_{position} + \mathbf{E}_{layout} + \mathbf{E}_{vision} \nonumber
\end{equation}



### 4.2 Multitask Header
In BROS++, Multitask Learning handles RE and EE simultaneously. Both transformer heads have the same input. First Token Classification (ITC) and Sequence Token Classification (STC) are EE's labeling tasks. ITC identifies the phrase's first token, whereas STC anticipates nearby tokens. The Spatial Dependency Parsing (SPADE) header extracts layout links in RE. Layout connections are separated into horizontal and vertical groups by link direction. Two SPADE headers are applied, one for each group, since it can identify only one box with a connection to a box, whereas datasets may have a maximum of 2 keys connecting to a value.

### 4.3 Text Augmentation
In the field of document understanding, data deficiency is a prevalent problem. We suggest introducing Text Replacement and Random Loss Characters to increase data diversity. TR replaces similar text in the layout area. RLC randomly loses text characters (including space characters). The text results from an OCR model, and it is hard to build a deep-learning model forecast with 100 percent accuracy. Therefore, RLC is advised to construct BROS++, which can effectively anticipate common scenarios and extreme instances where the OCR model incorrectly guesses particular characters.

## 5. Experiments

### 5.1 Experiment setting

**Dataset**: We run experiments on VinEHR and three bench- mark datasets: FUNSD [10], SROIE [11] and SciTSR [12]. They are mentioned and described in Section 2 and Section 3. 

**Implementation Details**: The training, evaluating, testing, and inference phases of the repository are implemented by the Pytorch Lightning format and executed on an A100 GPU. We employ PhoBERT [16] to tokenize the Vietnamese VinEHR dataset. The model is trained with the Adam optimizer at 5 x 10−5. and linear decay. The batch size for VinEHR is set to 4 and for other benchmark datasets to 16. Experiments in the vision module determine the output size once RoI Align is performed (64, 4). The number of training epochs is as follows: 100 epochs for FUNSD and VinEHR, 30 for SROIE, and 7.5 for SciTSR
